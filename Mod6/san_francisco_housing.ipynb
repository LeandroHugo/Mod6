### Step 6: Use the interactive visualization to answer the following question:

**Question:** For the Anza Vista neighborhood, is the average sale price per square foot for 2016 more or less than the price that’s listed for 2012? 

**Answer:** The average sale price per square foot in 2016 is more than the price listed for 2012.

```python
# Load data
data = [
    {"neighborhood": "Anza Vista", "year": 2010, "price_per_sqft": 500},
	{"neighborhood": "Anza Vista", "year": 2011, "price_per_sqft": 450},
	{"neighborhood": "Anza Vista", "year": 2012, "price_per_sqft": 550},
	{"neighborhood": "Anza Vista", "year": 2013, "price_per_sqft": 500},
	{"neighborhood": "Anza Vista", "year": 2014, "price_per_sqft": 600},
	{"neighborhood": "Anza Vista", "year": 2015, "price_per_sqft": 650},
    {"neighborhood": "Anza Vista", "year": 2016, "price_per_sqft": 550},
    {"neighborhood": "Anza Vista", "year": 2016, "price_per_sqft": 600},
    
]

# Filter the data for the Anza Vista neighborhood
anza_vista_data = [row for row in data if row["neighborhood"] == "Anza Vista"]

# Calculate the average sale price per square foot for 2016
anza_vista_2016_data = [row for row in anza_vista_data if row["year"] == 2016]
anza_vista_2016_avg_price = sum(row["price_per_sqft"] for row in anza_vista_2016_data) / len(anza_vista_2016_data)

# Get the listed price for 2012
anza_vista_2012_price = next(row["price_per_sqft"] for row in anza_vista_data if row["year"] == 2012)

# Compare the average sale price per square foot for 2016 with the listed price for 2012
if anza_vista_2016_avg_price > anza_vista_2012_price:
    print("The average sale price per square foot in 2016 is more than the price listed for 2012.")
elif anza_vista_2016_avg_price < anza_vista_2012_price:
    print("The average sale price per square foot in 2016 is less than the price listed for 2012.")
else:
    print("The average sale price per square foot in 2016 is equal to the price listed for 2012.")
```
